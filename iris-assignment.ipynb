# 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

```python
#ploting a heatmap to show the correlation between variables
sns.heatmap(iris_data.corr(), linecolor = 'blue', linewidths = 1)
```

```python
#correlation between variables are annotated
sns.heatmap(iris_data.corr(), linecolor = 'blue', linewidths = 1,annot=True)
```

```python
#pair plot of the variables is plotted
pair = sns.pairplot(iris_data)
pair
```

```python
#The classification of iris species based on Petal length is shown
sns.swarmplot(x="Classification",y="PL",data=iris_data)
```

```python
#Identifying median,quartiles and outliers using boxplot
sns.boxplot(x="PL",y="Classification",data=iris_data)
```

```python
#Scatter plot considering Sepal length and Sepal Width
iris_data.plot(kind="scatter", x="SL", y="SW")
```
